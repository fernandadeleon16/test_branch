```python
#Se hará eñ mismo tratamiento que se hizo en lecciones anteriores, para 
#generar la duración de cada registro 
#Como primer paso, se añadirá un nuevo campo de duración, para no modificar la estructura original

df['duration_num'] = df['duration']
```

```python
#Además se generará un campo adicional de unidades, para guardar la información, que podrá ser 'min' o 'season'.
df['duration_unit'] = ''
```

```python
#Se verifica el nuevo campo
df.head(3)
```

```python
#Como primer paso, se cambia el tipo de columna de las columnas de duration
df['duration_unit'] = df['duration_unit'].astype(pd.StringDtype())
```

```python
df.dtypes
```

```python
#Si lo quisiera hacer para más columnas a la vez (todas las de tipo object)
#Podría primero buscar todas las columnas de ese tipo 
df.select_dtypes(include='object').columns

#Y luego aplicarlo
```

```python
#Cambia el tipo de variable de todas las columnas tipo object 
stringcols = df.select_dtypes(include='object').columns
df[stringcols] = df[stringcols].fillna('').astype(pd.StringDtype())
```

```python
df.dtypes
```

```python
df.sample(3)
```

```python
#Luego, se buscará los registros que tengan 'min', y con ellos, se asignará valores 
#a la column duration_unit

#El primer paso es definir a todas las unidades como 'min', luego esto se cambiará
df['duration_unit'] = 'min'
```

```python
df.head(3)
```

```python
#Usando where asingamos la duration_unit a min cuando es 'min' y a 'season' caundo es <> min 
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.where.html
df['duration_unit'] = df['duration_unit'].where(df.duration_num.str.endswith('min'), 'season')
```

```python
df.sample(5)
#Se puede ver parte de un dataframe que no sea el head o tail con sample
```

```python
#Se usa el método replace para el reemplaso de los strings 
#En este caso se reemplaza todas as ocurrencias de min, Seasons por blancos 
df['duration_num'] = df['duration_num'].replace({'min':'', 'Season':'','s':''}, regex=True)
```

```python
df.sample(5)
```
